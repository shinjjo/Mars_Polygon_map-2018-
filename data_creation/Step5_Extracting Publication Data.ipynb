```python
cno = 0
papers = []
for i in data:
    checker = i['properties']['number_of_papers']
    cno += 1
    if checker != 0:
        q = ads.SearchQuery(q=i['properties']['name'], keyword='mars', database='astronomy', fl=['bibcode', 'title', 'author', 'year', 'pub', 'ciation_count'], rows = 2000)
        for p in q:
            article = [i['properties']['id'], i['properties']['name'], p.title, p.author, p.year, p.bibcode, p.pub, p.citation_count]
            papers.append(article)
        print(str(cno) + str(i['properties']['name']) + " info added ")
    else:
        pass
```
