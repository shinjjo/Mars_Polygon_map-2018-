```python
import json
```

```python
#Converting cordinates of [WFS.json] file
#0 to 360 -> -180 to 180 
#from https://astrocloud.wr.usgs.gov/dataset/data/nomenclature/MARS/WFS?request=getFeature&service=WFS&version=1.1.0&outputformat=application/json 
```

```python
with open('mercury_init.json') as fd:
     json_data = json.load(fd)
```

```python
'''
Feature Coordinates
mars_init.json
venus_init.json
moon_init.json
mercury_init.json
'''
```

```python
# Converting from [0 to 360] to [-180 to 180]
numbers = len(json_data['features'])
for i in range(numbers):
    ex = json_data['features'][i]['geometry']['coordinates']
    check = json_data['features'][i]['geometry']['type']
    if check == 'MultiPolygon':
        for j in range(len(ex)):
            for k in range(len(ex[j][0])):
                if ex[j][0][k][0] > 180:
                    json_data['features'][i]['geometry']['coordinates'][j][0][k][0] = ex[j][0][k][0]-360                    
    elif check == 'Polygon' or check == 'MultiLineString':
        for j in range(len(ex)):
            for k in range(len(ex[j])):
                if ex[j][k][0] > 180:
                    json_data['features'][i]['geometry']['coordinates'][j][k][0] = ex[j][k][0]-360
    elif check == 'Point':
        if ex[0] > 180:
            json_data['features'][i]['geometry']['coordinates'][0] = ex[0]-360
    elif check == 'LineString':
        for j in range(len(ex)):
            if ex[j][0] > 180:
                json_data['features'][i]['geometry']['coordinates'][j][0] = ex[j][0]-360
```

```python
#Delete unapproved rows
for i in range(len(coord)):
    try:
        if coord[i]['properties']['Approval_Status'] != 'Approved':
            del coord[i]
    except:
        pass
```
